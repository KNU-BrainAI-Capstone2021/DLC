```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('musinsa_test.csv', encoding='utf-8-sig')
```

```python
df
```

```python
coat=df['outer'].str.count('코트' or 'coat' or 'TRENCH')
jacket=df['outer'].str.count('자켓' or 'JACKET' or '블루종')
blazer=df['outer'].str.count('블레이저' or 'blazer')
```

```python
coat_num=coat.sum()
jacket_num=jacket.sum()
blazer_num=blazer.sum()
```

```python
num={'coat':[coat_num],'jacket':[jacket_num],'blazer':[blazer_num]}
num_df=pd.DataFrame(num)
plt.bar(num_df.columns,num_df.iloc[0])
```

```python
shirts=df['top'].str.count('셔츠' or 'shirts')
knit=df['top'].str.count('니트' or 'knit')
mantoman=df['top'].str.count('맨투맨')
gadigan=df['top'].str.count('가디건')
sleeve=df['top'].str.count('슬리브')
```

```python
shirts_num=shirts.sum()
knit_num=knit.sum()
mantoman_num=mantoman.sum()
gadigan_num=gadigan.sum()
sleeve_num=sleeve.sum()
```

```python
num={'shirts':[shirts_num],'knit':[knit_num],'mantoman':[mantoman_num],'gadigan':[gadigan_num],'sleeve':[sleeve_num]}
num_df=pd.DataFrame(num)
plt.bar(num_df.columns,num_df.iloc[0])
```
